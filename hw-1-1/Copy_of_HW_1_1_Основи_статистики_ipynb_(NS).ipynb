–í —Ü—å–æ–º—É –¥–æ–º–∞—à–Ω—å–æ–º—É –∑–∞–≤–¥–∞–Ω–Ω—ñ –≤—ñ–¥–ø—Ä–∞—Ü—é—î–º–æ –∑ –≤–∞–º–∏ –∑–Ω–∞–Ω–Ω—è, –æ—Ç—Ä–∏–º–∞–Ω—ñ –≤ –º–æ–¥—É–ª—ñ –∑—ñ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏.

**–ó–∞–≤–¥–∞–Ω–Ω—è 1.** –ú–∏ –æ–±–≥–æ–≤–æ—Ä–∏–ª–∏ –∑ –í–∞–º–∏ –Ω–∞ –ª–µ–∫—Ü—ñ—ó, —â–æ –≤–µ–ª–∏–∫–∞ –ª—ñ—Ç–µ—Ä–∞ –°—ñ–≥–º–∞ –≥—Ä–µ—Ü—å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤—ñ—Ç—É –ø–æ–∑–Ω–∞—á–∞—î –≤ –º–∞—Ç–µ–º–∞—Ç–∏—Ü—ñ —Å—É–º—É. –ó–Ω–∞–π–¥—ñ—Ç—å –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º Python $\Sigma x$, $\Sigma x^2$, —Ç–∞ $\Sigma (x-1)^2$ –¥–ª—è –Ω–∞—Å—Ç—É–ø–Ω–æ–≥–æ –Ω–∞–±–æ—Ä—É –¥–∞–Ω–∏—Ö:
`1 3 4`
–í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∫–æ–∂–Ω–æ–≥–æ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –Ω–∞ –µ–∫—Ä–∞–Ω.

```python
data1 = [1, 3, 4]
sum_data1 = sum(data1)
sum_data1_x2 = sum([x**2 for x in data1])
sum_data1_min1 = sum([(x-1)**2 for x in data1])
print('Œ£ùë•===', sum_data1)
print('Œ£ùë•2===',sum_data1_x2)
print('Œ£(ùë•‚àí1)2===',sum_data1_min1)
```

**–ó–∞–≤–¥–∞–Ω–Ω—è 2.** –ú–∏ –∑—ñ–±—Ä–∞–ª–∏ –¥–∞–Ω—ñ –ø—Ä–æ —Å—É–º—É –ø–æ–∫—É–ø–æ–∫ –≤ –≥—Ä–∏–≤–Ω—è—Ö 20–º–∞ –≤–∏–ø–∞–¥–∫–æ–≤–æ –æ–±—Ä–∞–Ω–∏–º–∏ –∫–ª—ñ—î–Ω—Ç–∞–º–∏ –≤ –Ω–∞—à–æ–º—É –º–∞–≥–∞–∑–∏–Ω—ñ (–≤—Å—å–æ–≥–æ –∫–ª—ñ—î–Ω—Ç—ñ–≤ —É –Ω–∞—Å –±—ñ–ª—å—à–µ) –∑–∞ –º–∏–Ω—É–ª–∏–π –º—ñ—Å—è—Ü—å. –í–æ–Ω–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ñ –≤ –∑–º—ñ–Ω–Ω—ñ–π `sum_purchases`. –ó–Ω–∞–π–¥—ñ—Ç—å **–≤–∏–±—ñ—Ä–∫–æ–≤–µ —Å–µ—Ä–µ–¥–Ω—î** –¥–ª—è —Ü–∏—Ö –¥–∞–Ω–∏—Ö –∫–æ–∂–Ω–∏–º –∑ –∑–∞–∑–Ω–∞—á–µ–Ω–∏—Ö –º–µ—Ç–æ–¥—ñ–≤:   

  2.1. –±–µ–∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –∂–æ–¥–Ω–∏—Ö –±—ñ–±–ª—ñ–æ—Ç–µ–∫, –∫–æ—Ä–∏—Å—Ç—É—é—á–∏—Å—å —Ç—ñ–ª—å–∫–∏ Python.  
  2.2. –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π –º–µ—Ç–æ–¥ –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ `numpy`.

  –í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—á–∏—Å–ª–µ–Ω—å –≤ –∫–æ–∂–Ω–æ–º—É –∑ –≤–∏–ø–∞–¥–∫—ñ–≤ –Ω–∞ –µ–∫—Ä–∞–Ω —Ç–∞ –ø–æ—Ä—ñ–≤–Ω—è–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏. –í–æ–Ω–∏ —Å–ø—ñ–≤–ø–∞–¥–∞—é—Ç—å?

```python
sum_purchases = [ 316,  493,  367,  153,  224,  121,  322,  450,  196,  397,  281,
  239,  484,  124,  223,  448,  432,  255, 4310, 3524]
```

```python
mean_py = sum(sum_purchases)/len(sum_purchases)
print('–≤–∏–±—ñ—Ä–∫–æ–≤–µ —Å–µ—Ä–µ–¥–Ω—î PY:', mean_py)
```

```python
import numpy as np

mean_np = np.mean(sum_purchases)
print('–≤–∏–±—ñ—Ä–∫–æ–≤–µ —Å–µ—Ä–µ–¥–Ω—î NP:', mean_np)
```

**–ó–∞–≤–¥–∞–Ω–Ω—è 3.** –î–ª—è —Ç–æ–≥–æ –∂ –Ω–∞–±–æ—Ä—É –¥–∞–Ω–∏—Ö –≤ –∑–º—ñ–Ω–Ω—ñ–π `sum_purchases` –æ–±—á–∏—Å–ª—ñ—Ç—å **–º–µ–¥—ñ–∞–Ω—É** –∫–æ–∂–Ω–∏–º –∑ –∑–∞–∑–Ω–∞—á–µ–Ω–∏—Ö –º–µ—Ç–æ–¥—ñ–≤:  

  3.1. –±–µ–∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –∂–æ–¥–Ω–∏—Ö –±—ñ–±–ª—ñ–æ—Ç–µ–∫, –∫–æ—Ä–∏—Å—Ç—É—é—á–∏—Å—å —Ç—ñ–ª—å–∫–∏ Python.  
  3.2. –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π –º–µ—Ç–æ–¥ –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ numpy.  

–í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—á–∏—Å–ª–µ–Ω—å –≤ –∫–æ–∂–Ω–æ–º—É –∑ –≤–∏–ø–∞–¥–∫—ñ–≤ –Ω–∞ –µ–∫—Ä–∞–Ω. –ü–æ—Ä—ñ–≤–Ω—è–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ –∑—ñ –∑–Ω–∞—á–µ–Ω–Ω—è–º –≤–∏–±—ñ—Ä–∫–æ–≤–æ–≥–æ —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ. –ù–∞–ø–∏—à—ñ—Ç—å –∫–æ–º–µ–Ω—Ç–∞—Ä, –¥–ª—è —è–∫–∏—Ö –∑–∞–¥–∞—á –≤–∏ –± –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞–ª–∏ –≤–∏–±—ñ—Ä–∫–æ–≤–µ —Å–µ—Ä–µ–¥–Ω—î, –∞ –¥–ª—è —è–∫–∏—Ö - –º–µ–¥—ñ–∞–Ω—É?

```python
median_py = sorted(sum_purchases)[len(sum_purchases)//2] + sorted(sum_purchases)[len(sum_purchases)//2-1]
median_py = median_py/2
print('–º–µ–¥—ñ–∞–Ω–∞ PY:', median_py)
```

```python
median_np = np.median(sum_purchases)
print('–º–µ–¥—ñ–∞–Ω–∞ NP:', median_np)
```

**–í–ò–°–ù–û–í–û–ö**

*–û—Å–∫—ñ–ª—å–∫–∏ –≤ –¥–∞–Ω—ñ–π –≤–∏–±—ñ—Ä—Ü—ñ —î –¥–≤–∞ –¥—É–∂–µ –≤–µ–ª–∏–∫–∏—Ö —á–∏—Å–ª–∞, —è–∫—ñ —Å–∏–ª—å–Ω–æ –≤–∏–±–∏–≤–∞—é—Ç—å—Å—è –∑ –∑–∞–≥–∞–ª—å–Ω–æ—ó –∫–∞—Ä—Ç–∏–Ω–∏, —Ç–æ —Å–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –¥—É–∂–µ –ø–µ—Ä–µ—Ç—è–≥—É—î—Ç—å—Å—è –≤–≥–æ—Ä—É —ñ –Ω–µ –≤—ñ–¥–æ–±—Ä–∞–∂–∞—î –¥–æ–±—Ä–µ –±—ñ–ª—å—à—ñ—Å—Ç—å –∑–Ω–∞—á–µ–Ω—å. –ü–æ —Å—É—Ç—ñ –≤—Å—ñ —á–µ–∫–∏, –æ–∫—Ä—ñ–º —Ü–∏—Ö –¥–≤–æ—Ö –≤–∏–∫–∏–¥—ñ–≤ –º–µ–Ω—à—ñ –∑–∞ —Å–µ—Ä–µ–¥–Ω—î.
–¢–∞–∫–∏–º —á–∏–Ω–æ–º –º–µ–¥—ñ–∞–Ω–∞ –≤–∏—Ö–æ–¥–∏—Ç—å –∑–Ω–∞—á–Ω–æ —ñ–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω—ñ—à–æ—é.
–£ –≤–∏–ø–∞–¥–∫—É –±—ñ–ª—å—à –æ–¥–Ω–æ—Ä—ñ–¥–Ω–æ–≥–æ —Ä–æ–∑–ø–æ–¥—ñ–ª—É –¥–∞–Ω–∏—Ö –≤–∏–±—ñ—Ä–∫–æ–≤–µ —Å–µ—Ä–µ–¥–Ω—î —Å–ø—Ä–∞—Ü—é–≤–∞–ª–æ –± –∫—Ä–∞—â–µ.*

**–ó–∞–≤–¥–∞–Ω–Ω—è 4**. –û–±—á–∏—Å–ª—ñ—Å—Ç—å —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è –Ω–∞–≤–µ–¥–µ–Ω–æ—ó –≤–∏–±—ñ—Ä–∫–∏ –≤ –∑–º—ñ–Ω–Ω—ñ–π `sum_purchases` –∫–æ–∂–Ω–∏–º –∑ –∑–∞–∑–Ω–∞—á–µ–Ω–∏—Ö –º–µ—Ç–æ–¥—ñ–≤:  

  4.1. –±–µ–∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –∂–æ–¥–Ω–∏—Ö –±—ñ–±–ª—ñ–æ—Ç–µ–∫, –∫–æ—Ä–∏—Å—Ç—É—é—á–∏—Å—å —Ç—ñ–ª—å–∫–∏ Python.  
  4.2. –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π –º–µ—Ç–æ–¥ –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ numpy.  

–í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—á–∏—Å–ª–µ–Ω—å –≤ –∫–æ–∂–Ω–æ–º—É –∑ –≤–∏–ø–∞–¥–∫—ñ–≤ –Ω–∞ –µ–∫—Ä–∞–Ω.

```python
data1_minus_mean_sqr = [(x - mean_py)**2 for x in sum_purchases]
sample_std_py = (sum(data1_minus_mean_sqr)/(len(sum_purchases)-1))**0.5
print('–≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è PY:', sample_std_py)
```

```python
sample_std_np = np.std(sum_purchases, ddof=1)
print('–≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è NP:', sample_std_np)
```

**–ó–∞–≤–¥–∞–Ω–Ω—è 5**. –ú–∏ –ø—Ä–æ–≤–µ–ª–∏ –¥–æ—Å–ª—ñ–¥–∂–µ–Ω–Ω—è –≤—ñ–∫—É —Å—Ç—É–¥–µ–Ω—Ç—ñ–≤-–ø–µ—Ä—à–æ–∫—É—Ä—Å–Ω–∏–∫—ñ–≤. –í–∏—è–≤–∏–ª–æ—Å—å, —â–æ
- 10 —Å—Ç—É–¥–µ–Ω—Ç—ñ–≤ –º–∞—é—Ç—å –≤—ñ–∫ 16 —Ä–æ–∫—ñ–≤
- 25 —Å—Ç—É–¥–µ–Ω—Ç—ñ–≤ - 17 —Ä–æ–∫—ñ–≤
- 20 —Å—Ç—É–¥–µ–Ω—Ç—ñ–≤ - 18 —Ä–æ–∫—ñ–≤
- 2 —Å—Ç—É–¥–µ–Ω—Ç–∏ - 20 —Ä–æ–∫—ñ–≤

–û–±—á–∏—Å–ª—ñ—Å—Ç—å —Å–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –¥–ª—è —Ü—ñ—î—ó –≤–∏–±—ñ—Ä–∫–∏ –±—É–¥—å-—è–∫–∏–º —Å–ø–æ—Å–æ–±–æ–º - –Ω–∞ —á–∏—Å—Ç–æ–º—É Python –∞–±–æ –∑ numpy - —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –µ–∫—Ä–∞–Ω.

```python
ages = [16]*10 + [17]*25 + [18]*20 + [20]*2
ages_mean_py = sum(ages)/len(ages)
ages_mean_np = np.mean(ages)
print('–°–µ—Ä–µ–¥–Ω—î PY:', ages_mean_py)
print('–°–µ—Ä–µ–¥–Ω—î NP:', ages_mean_np)
```

**–ó–∞–≤–¥–∞–Ω–Ω—è 6**. –ö—Ä—ñ–º –¥–∞–Ω–∏—Ö –ø—Ä–æ —Å—É–º—É –ø–æ–∫—É–ø–æ–∫, –º–∏ –∑—ñ–±—Ä–∞–ª–∏ —â–µ –¥–∞–Ω—ñ –ø—Ä–æ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤—ñ–¥–≤—ñ–¥—É–≤–∞–Ω—å –∫–æ–∂–Ω–æ–≥–æ –∑ –∫–æ–∂–Ω–æ–≥–æ –ø–æ–∫—É–ø—Ü—è, —è–∫—ñ –∑–Ω–∞—Ö–æ–¥—è—Ç—å—Å—è –≤ –∑–º—ñ–Ω–Ω—ñ–π `visits`. –û–±—á–∏—Å–ª—ñ—Ç—å –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç –∫–æ—Ä–µ–ª—è—Ü—ñ—ó –ü—ñ—Ä—Å–æ–Ω–∞ –¥–ª—è —Å—É–º–∏ –ø–æ–∫—É–ø–æ–∫ —ñ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≤—ñ–¥–≤—ñ–¥—É–≤–∞–Ω—å –∫–æ–∂–Ω–∏–º –∑ –∑–∞–∑–Ω–∞—á–µ–Ω–∏—Ö –º–µ—Ç–æ–¥—ñ–≤:  

  6.1. –±–µ–∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –∂–æ–¥–Ω–∏—Ö –±—ñ–±–ª—ñ–æ—Ç–µ–∫, –∫–æ—Ä–∏—Å—Ç—É—é—á–∏—Å—å —Ç—ñ–ª—å–∫–∏ Python.  
  6.2. –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π –º–µ—Ç–æ–¥ –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ numpy.  
  
–í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –µ–∫—Ä–∞–Ω. –ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

–í–∏–≤–µ–¥—ñ—Ç—å –¥–æ–¥–∞—Ç–∫–æ–≤–æ –≥—Ä–∞—Ñ—ñ–∫ —Ä–æ–∑—Å—ñ—é–≤–∞–Ω–Ω—è –≤–∏–∫–ª–∏–∫–∞–≤—à–∏ –∫–æ–¥ –Ω–∏–∂—á–µ —ñ –∑—Ä–æ–±—ñ—Ç—å –∑–∞–∫–ª—é—á–µ–Ω–Ω—è, —á–∏ –≤–∞—Ä—Ç–æ —â–æ—Å—å –∑–º—ñ–Ω–∏—Ç–∏ –ø—Ä–∏ –ø—ñ–¥—Ä–∞—Ö—É–Ω–∫—É –∫–æ—Ä–µ–ª—è—Ü—ñ—ó, –∞–±–∏ –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç –ü—ñ—Ä—Å–æ–Ω–∞ –ø–æ–∫–∞–∑—É–≤–∞–≤ –∫–æ—Ä–µ–∫—Ç–Ω—É –æ—Ü—ñ–Ω–∫—É –ª—ñ–Ω—ñ–π–Ω–æ—ó –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ?

```python
visits = [ 6,  2, 10,  3,  9,  9,  2,  3,  6,  6, 10,  2,  5, 10,  3,  5,  4,
        7,  4,  2]
```

```python
mean_pur = sum(sum_purchases) / len(sum_purchases)
mean_vis = sum(visits) / len(visits)

print("mean_pur:", mean_pur)
print("mean_vis:", mean_vis)

xy = [(sum_purchases[i] - mean_pur) * (visits[i] - mean_vis)
      for i in range(len(sum_purchases))]

cov_pur_vis = sum(xy) / (len(sum_purchases) - 1)
print("cov_pur_vis:", cov_pur_vis)


# –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è –ø–æ sum_purchases
pur_minus_mean_sqr = [(x - mean_pur)**2 for x in sum_purchases]
std_pur = (sum(pur_minus_mean_sqr) / (len(sum_purchases) - 1)) ** 0.5

# –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è –ø–æ visits
vis_minus_mean_sqr = [(v - mean_vis)**2 for v in visits]
std_vis = (sum(vis_minus_mean_sqr) / (len(visits) - 1)) ** 0.5

print("std_pur:", std_pur)
print("std_vis:", std_vis)

corr_py = cov_pur_vis / (std_pur * std_vis)
print("–ö–æ—Ä–µ–ª—è—Ü—ñ—è PY:", corr_py)
```

```python
corr_np = np.corrcoef(sum_purchases, visits)[0, 1]
print("–ö–æ—Ä–µ–ª—è—Ü—ñ—è NP:", corr_np)
```

**–ó–∞–≤–¥–∞–Ω–Ω—è 7**. –í –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ–º—É –∑–∞–≤–¥–∞–Ω–Ω—ñ –º–∏ –ø–æ–º—ñ—Ç–∏–ª–∏, —â–æ –≤–∏–∫–∏–¥–∏ –Ω–∞–º –∑–∞–≤–∞–∂–∞—é—Ç—å –æ—Ç—Ä–∏–º–∞—Ç–∏ "–ø—Ä–∞–≤–¥–∏–≤–µ" –∑–Ω–∞—á–µ–Ω–Ω—è –∫–æ—Ä–µ–ª—è—Ü—ñ—ó. –î–∞–≤–∞–π—Ç–µ –≤–∏–ª—É—á–∏–º–æ –∑–∞–ø–∏—Å–∏, —è–∫—ñ –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—é—Ç—å –∑–Ω–∞—á–µ–Ω–Ω—è–º –∑ –≤–∏–∫–∏–¥–∞–º–∏ –≤ —Å—É–º—ñ –ø–æ–∫—É–ø–æ–∫ —ñ –æ–±—á–∏—Å–ª—ñ—Ç—å –∫–æ–µ—Ñ. –∫–æ—Ä–µ–ª—è—Ü—ñ—ó. –î–ª—è —Ü—å–æ–≥–æ –ø–æ—Å—Ç–∞–≤—Ç–µ `threshold`, —è–∫–∏–π –¥–æ—Ä—ñ–Ω—é—î —Å–µ—Ä–µ–¥–Ω—å–æ–º—É –∑–Ω–∞—á–µ–Ω–Ω—é + 2 —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∏—Ö –≤–¥—ñ—Ö–∏–ª–ª—è —Å—É–º–∏ –ø–æ–∫—É–ø–æ–∫ —ñ –≤—Å—ñ –∑–Ω–∞—á–µ–Ω–Ω—è, —è–∫—ñ –±—ñ–ª—å—à–µ –∑–∞ –Ω—å–æ–≥–æ, –±—É–¥–µ–º–æ –≤–≤–∞–∂–∞—Ç–∏ –≤–∏–∫–∏–¥–∞–º–∏ (—î —ñ —ñ–Ω—à—ñ –ø—Ä–∞–≤–∏–ª–∞ –¥–ª—è –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è –≤–∏–∫–∏–¥—ñ–≤ - –∑ –Ω–∏–º –∏ –ø–æ–∑–Ω–∞–π–æ–º–∏–º–æ—Å—å –ø—ñ–∑–Ω—ñ—à–µ).

–í–∏–ª—É—á—ñ—Ç—å –∑–Ω–∞—á–µ–Ω–Ω—è –∑ –æ–±–æ—Ö –º–∞—Å–∏–≤—ñ–≤ —ñ –æ–±—á–∏—Å–ª—ñ—Ç—å –∫–æ–µ—Ñ. –∫–æ—Ä–µ–ª—è—Ü—ñ—ó –ü—ñ—Ä—Å–æ–Ω–∞ –∑ –¥–æ–ø–æ–º–æ–≥–æ—é `numpy` –¥–ª—è –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≤—ñ–∑–∏—Ç—ñ–≤ —ñ —Å—É–º–∏ –ø–æ–∫—É–ø–æ–∫ –±–µ–∑ –∑–∞–ø–∏—Å—ñ–≤-–≤–∏–∫–∏–¥—ñ–≤. –í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –µ–∫—Ä–∞–Ω, –ø–æ–±—É–¥—É–π—Ç–µ –¥—ñ–∞–≥—Ä–∞–º—É —Ä–æ–∑—Å—ñ—é–≤–∞–Ω–Ω—è —ñ –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç: —â–æ –º–æ–∂–µ–º–æ —Å–∫–∞–∑–∞—Ç–∏ –ø—Ä–æ –∑–∞–ª–µ–∂–Ω—ñ—Å—Ç—å –≤ –¥–∞–Ω–∏—Ö?

```python
mean_pur = np.mean(sum_purchases)
std_pur = np.std(sum_purchases, ddof=1)

threshold = mean_pur + 2 * std_pur
print("threshold:", threshold)

sum_purchases_clean = []
visits_clean = []

for s, v in zip(sum_purchases, visits):
    if s < threshold:
        sum_purchases_clean.append(s)
        visits_clean.append(v)

print("len original:", len(sum_purchases), "len clean:", len(sum_purchases_clean))

corr_clean = np.corrcoef(sum_purchases_clean, visits_clean)[0, 1]
print("–ö–æ—Ä–µ–ª—è—Ü—ñ—è –±–µ–∑ –≤–∏–∫–∏–¥—ñ–≤:", corr_clean)
```

```python
plt.scatter(sum_purchases_clean, visits_clean)
plt.title('Scatter Plot of Purchases vs. Visits (CLEAN)')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits');
```
